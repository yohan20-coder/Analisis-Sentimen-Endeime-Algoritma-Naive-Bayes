```python
# ============================
# 2. Import Library
# ============================
import pytesseract
from PIL import Image
import cv2
import pandas as pd

# ============================
# 3. Upload Screenshot
# ============================
from google.colab import files
uploaded = files.upload()

# Ambil nama file pertama yang diupload
image_path = list(uploaded.keys())[0]

# ============================
# 4. Preprocessing Gambar
# ============================
img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)[1]

# Simpan hasil preprocessing sementara
preprocessed_path = "processed.png"
cv2.imwrite(preprocessed_path, thresh)

# ============================
# 5. OCR dengan Tesseract
# ============================
text = pytesseract.image_to_string(Image.open(preprocessed_path), lang="eng+ind")

# ============================
# 6. Pisahkan Caption & Komentar
# ============================
lines = [line.strip() for line in text.split("\n") if line.strip() != ""]

caption = lines[0] if len(lines) > 0 else ""
komentar = lines[1:] if len(lines) > 1 else []

print("=== Caption ===")
print(caption)

print("\n=== Komentar ===")
for i, komen in enumerate(komentar, 1):
    print(f"{i}. {komen}")

# ============================
# 7. Simpan ke CSV
# ============================
df = pd.DataFrame({
    "caption": [caption] * max(1, len(komentar)),
    "komentar": komentar if komentar else [""]
})

df.to_csv("hasil_caption_komentar.csv", index=False)

# Download hasil
files.download("hasil_caption_komentar.csv")
```
